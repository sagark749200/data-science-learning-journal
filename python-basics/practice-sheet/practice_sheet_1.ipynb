### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
head = int(input('enter the heads value'))
legs = int(input('enter the legs value'))
chicken = str(input('enter the chicken value'))
dog = str(input('enter the dog value'))

dog = (legs - 2*head)/2   #mathematical expression
chicken = head - dog

print(chicken,dog)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
first_term = int(input('enter the number'))
second_term = int(input('enter the number'))
n = int(input('enter the value of n'))

d = second_term - first_term
an = first_term + (n-1)*d

print(an)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
n1 = int(input('enter num1'))
d1 = int(input('enter den1'))
n2 = int(input('enter num2'))
d2 = int(input('enter den2'))

sum = ((n1*d2) + (n2*d1))/(d1*d2)

print(sum)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
ht = float(input('enter the tank height'))
lt = float(input('enter the tank length'))
bt = float(input('enter the tank breadth'))

hg = float(input('enter the glass height'))
rg = float(input('enter the glass radius'))

tank_vol = ht*lt*bt
glass_vol = 3.14*hg*(rg**2)

no_of_glass = tank_vol/glass_vol

print(no_of_glass)
````
