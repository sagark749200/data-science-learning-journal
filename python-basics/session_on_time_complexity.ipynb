# **Practice on Time-Complexity Concept**

```python
# question - 1

number = int(input('enter the number'))

digits = '0123456789'
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10

print(result)

#Time complexity - log(n)
```

```python
# Question - 2

L = [1,2,3,4]

sum = 0
for i in L:
  sum = sum + i

product = 1
for i in L:
  product = product*i

print(sum,product)

#Time com. - O(n)
```

```python
# Question - 3

A = [1,2,3,4]
B = [5,6,7,8]
for i in A:
  for j in B:
    print(i,j)

# Time com - n*2
```

```python
# Question 5

L = [1,2,3,4,5]

for i in range(0,len(L)//2):
  other = len(L) - i -1
  temp = L[i]
  L[i] = L[other]
  L[other] = temp

print(L)

# time com - O(n)
```

```python
#Question 8

n = 345

sum = 0
while n>0:
  sum = sum + n%10
  n = n // 10

print(sum)

# time com - log n
```

```python
# Question 9

def fib(n):
  if n == 1 or n == 0:
    return 1
  else:
    return fib(n-1) + fib(n-2)

# time com - 2*n = exponential
```
